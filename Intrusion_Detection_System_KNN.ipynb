Title: Intrusion Detection System using K-Nearest Neighbors

Aim: To classify network activity as normal or intrusion using the KNN algorithm.

```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
df=pd.read_csv('/content/network_intrusion.csv')
df.head()
```

```python
df.columns = [
    'Flow_Duration',
    'Total_Fwd_Packets',
    'Total_Backward_Packets',
    'Fwd_Packet_Length_Max',
    'Fwd_Packet_Length_Min',
    'Flow_Bytes_s',
    'Flow_Packets_s',
    'Label_1',
    'Label_2',
    'Attack_Type'
]
```

```python
df.columns
```

```python
X_train = df[['Flow_Duration', 'Flow_Bytes_s']].values
y_train = df['Attack_Type'].values
```

```python
y_train = np.where(y_train == 'BENIGN', 'Normal', 'Intrusion')
y_train = np.where(y_train == 'DDoS', 'Intrusion', 'Normal')
```

```python
df['Attack_Type'].unique()
```

```python
df = df.dropna(subset=['Attack_Type'])
```

```python
y_train = np.where(df['Attack_Type'] == 'BENIGN',
                   'Normal',
                   'Intrusion')
```

```python
X_train = df[['Flow_Duration', 'Flow_Bytes_s']].values
```

```python
X_test = np.array([5000000, 10000000])
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))


def knn_classify(X_test, X_train, y_train, k=3):
    distances = []

    for x in X_train:
        dist = euclidean_distance(X_test, x)
        distances.append(dist)

    neighbors_index = np.argsort(distances)[:k]
    neighbors_labels = y_train[neighbors_index]

    from collections import Counter
    most_common = Counter(neighbors_labels).most_common(1)
    return most_common[0][0]
```

```python
prediction = knn_classify(X_test, X_train, y_train, k=3)
print("Predicted Result:", prediction)
```
